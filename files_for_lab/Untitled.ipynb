```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('/Users/iagogarrido/Documents/ironhack/unit1/lab-customer-analysis-round-5/files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
df
```

```python
X = df.drop('Total Claim Amount', axis=1)
Y = df['Total Claim Amount']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
Y_train
```

```python
numeric_cols = ['Customer Lifetime Value', 'Income', 'Months Since Policy Inception', 'Number of Open Complaints', 'Number of Policies']
X_train_numeric = X_train[numeric_cols]
X_test_numeric = X_test[numeric_cols]
```

```python
scaler = MinMaxScaler()
```

```python
X_train_normalized = scaler.fit_transform(X_train_numeric)
X_test_normalized = scaler.transform(X_test_numeric)
```

```python
X_train_normalized_df = pd.DataFrame(X_train_normalized, columns=numeric_cols)
X_test_normalized_df = pd.DataFrame(X_test_normalized, columns=numeric_cols)
```
